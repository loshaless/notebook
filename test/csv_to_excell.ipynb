```python
import pandas as pd
import json

# import txt file
with open('payload_disposition_code_qapm_20241207.txt', 'r') as f:
    lines = f.readlines()

if lines:
    # Remove leading/trailing whitespace and any newline characters
    data_string = lines[0].strip()

    # Manually replace escaped backslashes with actual backslashes
    data_string = data_string.replace('\\\\', '\\')
    # Manually replace escaped single quotes with actual single quotes
    data_string = data_string.replace('\\\'', '\'')
    # Manually replace escaped double quotes with actual double quotes
    data_string = data_string.replace('\\"', '"')

    try:
        # Load the list from the string, which is already a string representing list of json objects
        data = json.loads(data_string)

        # Print each JSON object with indent for better readability
        for item in data:
            print(json.dumps(item, indent=2))
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON: {e}")
        print("Raw data:")
        print(data_string)
else:
    print("The file is empty.")
```
