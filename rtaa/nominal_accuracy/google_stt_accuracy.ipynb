# GET DATA By Interaction ID

```python
from google.cloud import bigquery
from collections import defaultdict

client = bigquery.Client()

grouped_transcript = defaultdict(list)
def get_transcript_by_interaction(interaction_id):
    get_transcript_by_interaction = f"""
        SELECT
          interaction_id,
          speaker_type,
          transcription
        FROM
          `genai-avengers.cimbniagagenai_ws4a.con_ws4a_rtaa_tbl_transcription`
        WHERE interaction_id = '{interaction_id}'
        order by interaction_id, createdon asc;
    """
    query_job = client.query(get_transcript_by_interaction)
    row_iterator_of_transcript = query_job.result()

    for row in row_iterator_of_transcript:
        # grouped_transcript[interaction_id].append(row[2])
        grouped_transcript[interaction_id].append({
            "role": row[1],
            "message": row[2],
        })
```

```python
import numpy as np
from evaluate import load
cer = load("cer")

def convert_to_number(text):
    if pd.isna(text):
        return np.nan
    text = str(text)
    numbers = ''.join(char for char in text if char.isdigit())

    return numbers if numbers else np.nan

def calculate_cer(expected, result):
    if pd.isna(expected) or pd.isna(result):
        return np.nan

    predictions = [result]
    references = [expected]
    return float(cer.compute(predictions=predictions, references=references))
```
